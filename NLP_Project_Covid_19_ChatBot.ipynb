```python
data_queries = [q1,q2,q3,q4,q5,q6]
data = ["q1","q2","q3","q4","q5","q6"]
```

```python
# covidQuestionare = {}
def inputData(q):
  d = {}
  for i in range(len(q['Question'])):
    if not q['Question'][i] in d:
      d[q['Question'][i]]=q['Answer'][i]
  return d
```

```python
for i in range(len(data_queries)):
  data[i] = inputData(data_queries[i])
```

###Preprocessing

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Defining the preprocessing of a Key and a Value in a Dictionary
def preprocessQuery(q):
  doc = nlp(q)
  s = ""
  for j in doc:
    if j.pos_!="PUNCT" and nlp.vocab[j.text].is_stop==False:
      s = "{0} {1}".format(s,j.lemma_)
  return s
```
